We can describe and model the change in position of a mass connected to a spring after being streched or compressed over time using the second order differential equation \begin{equation*}
mx'' = -kx
\end{equation*}
Where $m$ is the mass of the connected object, $x$ is the positional displacement of the connected object and $k$ is the spring constant



---


This equation can be derived by 
combining the principles behind Hooke's Law and Newtons second law of motion as follows:

Hooke's Law: 

\begin{equation*}
F = -kx
\end{equation*}

Newton's Second Law of Motion: 

\begin{equation*}
F = ma
\end{equation*}




First we can put Newton's Second Law of Motion in to terms of $x$ as follows, since acceleration ($a$) is the second derivative of displacement

\begin{equation*}
F = m \frac{d^2 x}{dt^2} = m x''
\end{equation*}


From here we can substitute $F$ from Newtons Second Law for $-kx$ in Hooke's Law to finally produce the differential equation

\begin{equation*}
mx'' = F =  -kx
\end{equation*}

Thus,

\begin{equation*}
mx'' =  -kx
\end{equation*}

We can also rewrite this as,

\begin{equation*}
mx'' + kx = 0
\end{equation*}




---


For example, lets say we have a spring-mass system described by the equation above where $m=1$ and $k=1$.  Additionally, the initial position $x = 0$ and the initial velocity $x' = 1$ for the mass.  This will give the initial starting conditions:

\begin{align*}
1x'' + 1x & =  0\\
x(0) & =  0\\
x'(0) & =  1
\end{align*}

We can find the value for $x(t)$ that will work as a general solution for $x''(t) = - x(t)$ using sympy as shown below:

```python
from sympy import *
```

```python
t = symbols('t')
x = symbols('x', cls=Function)
```

```python
expression = Eq(x(t).diff(t, t), -x(t))
expression
```

```python
dsolve(expression,x(t))
```

We find that

\begin{equation*}
x(t) = C_1 \cos t + C_2 \sin t.
\end{equation*}

Is a working solution 

If we plug in $C_1 = 0$ and $C_2 = 1$ based on our initial condition requirements of,

\begin{align*}
x(0) & =  0\\
x'(0) & =  1
\end{align*}

We will get the final harmonic oscillating function

\begin{equation*}
x(t) = \sin t.
\end{equation*}

We can then plot this function as shown below showing the position of the mass with respect to time.


```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x_values = np.linspace(0,30,10000)
y_values = np.sin(x_values)
plt.plot(x_values, y_values)
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Undamped spring-mass system")
```

The above example was a demonstration of an undamped spring-mass system however we can add a damping force proportional to the velocity of the mass in the system to simulate resistance.

This creates an additional force,

\begin{equation*}
F = -b x'
\end{equation*}

We can add this additional force to our differential equation we previously used to model the spring-mass system and produce a new differential equation,

\begin{equation*}
mx'' = -bx' - kx
\end{equation*}

Which can be rewritten as,

\begin{equation*}
mx'' + bx' + kx = 0
\end{equation*}

To model damped spring-mass system.


For example, lets say we have a spring-mass system described by the equation,

\begin{equation*}
x'' + 3x' + 2x = 0.
\end{equation*}

Using the format described above to model a damped spring-mass system we can conclude that $m = 1, b = 3$ and $k = 2$

Additionally, we can find the value for $x(t)$ that will work as a general solution for $x(t)'' + 3x(t)' + 2x(t) = 0.$ using sympy as shown below:

```python
expression = Eq(x(t).diff(t, t) + 3*x(t).diff(t) + 2*x(t), 0)
expression
```

```python
xt = dsolve(expression,x(t))
xt
```

If the initial position $x = 0$ and the initial velocity $x' = 1$ for the mass.  This will give the initial starting conditions:

\begin{align*}
x(0) & =  0\\
x'(0) & =  1
\end{align*}

Since we get

\begin{equation*}
x(t) = C_1 e^{-t} + C_2 e^{-2t}
\end{equation*}

And, 

\begin{equation*}
x'(t) = -C_1 e^{-t} -2 C_2 e^{-2t}
\end{equation*}

We can conclude from the initial conditions that

\begin{align*}
C_1 + C_2 & = 0\\
-C_1 -2C_2 & = 1
\end{align*}

Therefore $C_1 = 1$ and $C_2 = -1$ giving us the final equation for the damped spring-mass system as,

\begin{equation*}
x(t) =  e^{-t} - e^{-2t}
\end{equation*}

We can then plot this function as shown below showing the position of the mass with respect to time.





```python
x_values = np.linspace(0,10,10000)
y_values = np.exp(-1 * x_values) - np.exp(-2 * x_values)
plt.plot(x_values, y_values)
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Over-damped spring-mass system")
```



---


The spring-mass system above is over-damped since the damping cancels out any oscillation, however if we have what is known as an under-damped spring-mass system, the mass would still oscillate, but the oscillations would progressivly become smaller over time.

We can demonstrate this using a spring-mass system with the initial value problem,

\begin{align*}
x'' + 2x' + 50 x & = 0\\
x(0) & = 0\\
x'(0) & = 1
\end{align*}

We can find the value for $x(t)$ that will work as a general solution for $x'' + 2x' + 50x = 0 $ using sympy as shown below:

```python
expression = Eq(x(t).diff(t, t) + 2*x(t).diff(t) + 50*x(t), 0)
expression
```

```python
xt = dsolve(expression,x(t))
xt
```

If the initial position $x = 0$ and the initial velocity $x' = 1$ for the mass.  We know the initial starting conditions are:

\begin{align*}
x(0) & =  0\\
x'(0) & =  1
\end{align*}

We can conclude from the initial conditions that $C_1 = \frac{1}{7}$ and $C_2 = 0$ giving us the final equation for the damped spring-mass system as,

\begin{equation*}
x(t) = \frac{1}{7} e^{-t} \sin 7t
\end{equation*}

We can then plot this function as shown below showing the position of the mass with respect to time.

```python
x_values = np.linspace(0,10,10000)
y_values = 1/7 * np.sin(7 * x_values) * np.exp(-1 * x_values)
plt.plot(x_values, y_values)
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Under-damped spring-mass system")
```
